```python
import pandas as pd
import re
```

```python
historias_usuario = [
    "Como usuário, eu quero clicar no ícone do Simple Text para iniciar o aplicativo facilmente.",
    "Como usuário, eu quero organizar minhas notas de estudo utilizando campos de texto por disciplina para facilitar o acesso e a revisão das informações.",
    "Como usuário, eu quero usar o aplicativo para registrar minhas despesas diárias utilizando campos de texto para ajudar na gestão financeira pessoal e no controle de gastos.",
    "Como usuário, eu quero explorar as opções no menu de três pontos à direita para acessar funcionalidades adicionais como novo arquivo, salvar, salvar como, procurar, configurações e sair.",
    "Como usuário, eu quero selecionar a opção 'Salvar' para salvar meu documento atual e escolher o local adequado ou confirmar com os botões 'Save' ou 'Cancel'.",
    "Como usuário, eu quero garantir que não posso ter dois arquivos com o mesmo nome usando alerta nomes iguais para evitar problemas de sobrescrita involuntária.",
    "Como usuário, eu quero acessar a funcionalidade 'Abrir' para visualizar meus arquivos salvos com a opção de modificar ou excluir conforme necessário.",
    "Como usuário, eu quero utilizar a função de pesquisa (search) dentro do aplicativo para encontrar palavras específicas nos meus documentos.",
    "Como usuário, eu quero ter a opção de sair do aplicativo utilizando a função 'Exit' no menu de opções.",
    "Como usuário, eu quero selecionar 'Novo' no menu de três pontos para iniciar um novo arquivo de texto.",
    "Como usuário, eu quero acessar as configurações (Settings) no menu do aplicativo para ajustar preferências e configurações conforme minhas necessidades.",
    "Como usuário, eu quero usar a função 'Salvar Como' para salvar uma cópia modificada do meu arquivo atual com um novo nome ou em um local diferente.",
    "Como usuário, eu quero usar o Simple Text para tomar notas durante reuniões utilizando campos de texto para registrar claramente todos os pontos discutidos.",
    "Como usuário, eu quero usar o Simple Text para planejar detalhes da minha próxima viagem utilizando campos de texto para incluir itinerário, reservas e listas de itens a serem levados."
]
```

```python
regex_icone_simple_text = r"\bícone do Simple Text\b"
regex_campo_texto = r"\bcampo de texto\b"
regex_teclado = r"\bteclado\b"
regex_tres_pontinhos = r"\btrês pontinhos\b"
regex_nome_arquivo = r"\bnome do arquivo\b"
regex_lista_texto = r"\blista de texto\b"
regex_salvar = r"\bsalvar\b"
regex_cancelar = r"\bcancelar\b"
regex_alerta_nomes_iguais = r"\balerta nomes iguais\b"
regex_confirmar = r"\bconfirmar\b"
regex_descartar = r"\bdescartar\b"
regex_new = r"\bNovo\b"
regex_open = r"\bAbrir\b"
regex_save = r"\bSalvar\b"
regex_save_as = r"\bSalvar Como\b"
regex_search = r"\bsearch\b"
regex_setting = r"\bSettings\b"
regex_exit = r"\bExit\b"

mapeamento_campos = {
    "ícone do Simple Text": 'Simple Text Editor (accessibility id)',
    "campo de texto": 'com.maxistar.textpad:id/editText1',
    "teclado": 'android:id/decor_content_parent',
    "três pontinhos": 'More options',
    "nome do arquivo": 'com.maxistar.textpad:id/fdEditTextFile',
    "lista de texto": 'android:id/list',
    "salvar": 'com.maxistar.textpad:id/fdButtonCreate',
    "cancelar": 'com.maxistar.textpad:id/fdButtonCancel',
    "alerta nomes iguais": '/hierarchy/android.widget.FrameLayout',
    "confirmar": 'android:id/button1',
    "descartar": 'android:id/button2',
    "Novo": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[1]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "Abrir": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[2]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "Salvar": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[3]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "Salvar Como": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[4]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "search": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[5]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "Settings": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[6]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "Exit": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[7]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView'
}
```

```python
dataframes = []

for historia in historias_usuario:
    for campo, id_aplicacao in mapeamento_campos.items():
        match = None
        valor = None

        if "ícone do Simple Text" in campo:
            match = re.search(regex_icone_simple_text, historia)
        elif "campo de texto" in campo:
            match = re.search(regex_campo_texto, historia)
        elif "teclado" in campo:
            match = re.search(regex_teclado, historia)
        elif "três pontinhos" in campo:
            match = re.search(regex_tres_pontinhos, historia)
        elif "nome do arquivo" in campo:
            match = re.search(regex_nome_arquivo, historia)
        elif "lista de texto" in campo:
            match = re.search(regex_lista_texto, historia)
        elif "salvar" in campo:
            match = re.search(regex_salvar, historia)
        elif "cancelar" in campo:
            match = re.search(regex_cancelar, historia)
        elif "alerta nomes iguais" in campo:
            match = re.search(regex_alerta_nomes_iguais, historia)
        elif "confirmar" in campo:
            match = re.search(regex_confirmar, historia)
        elif "descartar" in campo:
            match = re.search(regex_descartar, historia)
        elif "Novo" in campo:
            match = re.search(regex_new, historia)
        elif "Abrir" in campo:
            match = re.search(regex_open, historia)
        elif "Salvar" in campo:
            match = re.search(regex_save, historia)
        elif "Salvar Como" in campo:
            match = re.search(regex_save_as, historia)
        elif "search" in campo:
            match = re.search(regex_search, historia)
        elif "Settings" in campo:
            match = re.search(regex_setting, historia)
        elif "Exit" in campo:
            match = re.search(regex_exit, historia)

        if match:
            valor = match.group(1).strip() if match.lastindex is not None and match.lastindex >= 1 else match.group().strip()

        if valor:
            dataframes.append(pd.DataFrame({
                "Activity": ["App activity"],
                "Field": [campo.split()[-1]],
                "ID": [id_aplicacao],
                "Action": ["click" if "botão" in campo else "type"],
                "Type": ["button" if "botão" in campo else "edittext"],
                "Size_Start": [1],
                "Size_End": [len(valor)],
                "Value": [valor]
            }))


if dataframes:
    df_final = pd.concat(dataframes, ignore_index=True)
    df_final.to_csv("SimpleText.csv", index=False)
else:
    pass
```
