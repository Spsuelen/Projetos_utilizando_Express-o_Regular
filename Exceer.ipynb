```python
import pandas as pd
import re
```

```python
historias_usuario = [
    "Como usuário, quero iniciar o aplicativo para iniciar minhas atividades fisicas",
    "Como usuário, quero iniciar o aplicativo Exceer e clicar no botão 'Start Now', para iniciar uma sessão de aquecimento dinâmico com uma contagem regressiva de 10 segundos.",
    "Como usuário durante o aquecimento dinâmico, quero ter a opção de pausar o cronômetro a qualquer momento, para poder retomar quando estiver pronto.",
    "Como usuário após concluir o aquecimento dinâmico, quero ser direcionado automaticamente para uma tela de agachamento nível 1 assistido, para realizar 4 séries de 4 sub-séries variando entre segundos e minutos de exercício.",
    "Como usuário durante o exercício de agachamento nível 1 assistido, desejo receber uma caixa de diálogo perguntando se desejo continuar para a próxima série, para poder escolher continuar ou encerrar o exercício.",
    "Como usuário após concluir o exercício de agachamento nível 1 assistido, quero ser questionado se desejo continuar com o próximo exercício de flexão de pernas, para poder escolher iniciar o próximo exercício ou encerrar a sessão.",
    "Como usuário durante o exercício de flexão de pernas, quero ter a opção de pausar o cronômetro a qualquer momento, para poder retomar quando estiver pronto.",
    "Como usuário durante o exercício de flexão de pernas, desejo ser questionado se desejo continuar com o próximo exercício de flexão de mãos ao final da terceira série, para poder escolher iniciar o próximo exercício ou encerrar a sessão.",
    "Como usuário após concluir o exercício de flexão de pernas, quero ser questionado se desejo continuar com o próximo exercício de flexão de mãos, para poder escolher iniciar o próximo exercício ou encerrar a sessão.",
    "Como usuário durante o exercício de flexão de parede, desejo ter um botão de 'Start' para iniciar cada série de quatro exercícios consecutivos, para iniciar a série de exercícios.",
    "Como usuário durante o exercício de flexão de parede, desejo receber uma caixa de diálogo perguntando se desejo continuar com a próxima série ao final da terceira série, para poder escolher continuar ou encerrar o exercício.",
    "Como usuário após concluir o exercício de flexão de parede, quero ser questionado se desejo continuar com o próximo exercício de horizontal pull vertical, para poder escolher iniciar o próximo exercício ou encerrar a sessão.",
    "Como usuário durante o exercício de horizontal pull vertical, desejo ter um botão de 'Start' para iniciar cada série de quatro exercícios consecutivos, para iniciar a série de exercícios.",
    "Como usuário durante o exercício de horizontal pull vertical, desejo receber uma caixa de diálogo perguntando se desejo continuar com a próxima série ao final da terceira série, para poder escolher continuar ou encerrar o exercício.",
    "Como usuário após concluir o exercício de horizontal pull vertical, quero ser questionado se desejo continuar com o próximo exercício de prancha ajoelhada, para poder escolher iniciar o próximo exercício ou encerrar a sessão.",
    "Como usuário durante o exercício de prancha ajoelhada, desejo ter um botão de 'Start' para iniciar cada série de exercícios, para iniciar a série de exercícios."
]
```

```python
regex_icone = r"Como usuário do (\w+),"
regex_start_now = r"iniciar o aplicativo Exceer e clicar no botão 'Start Now'"
regex_start = r"Exceer durante o (\w+)"
regex_exercicio = r"realizar (\d+) séries de (\d+) sub-séries"
regex_pause = r"pausar o cronômetro"
regex_series = r"ao final da (\w+) série"
regex_alerta = r"ser questionado se desejo continuar"
regex_botao_nao = r"botão 'Não'"
regex_botao_sim = r"botão 'Sim'"
regex_exercicio_dinamico = r"sessão de aquecimento dinâmico"
regex_agachamento = r"agachamento nível 1 assistido"
regex_flexao_pernas = r"flexão de pernas"
regex_flexao_maos = r"flexão de mãos"
regex_flexao_parede = r"flexão de parede"
regex_horizontal_pull = r"horizontal pull vertical"
regex_prancha = r"prancha ajoelhada"

mapeamento_campos = {
    "ícone": 'org.grapentin.apps.exceer.MainActivity',
    "start now": 'org.grapentin.apps.exceer:id/MainActivityTrainButton',
    "start": 'org.grapentin.apps.exceer:id/TrainingActivityContextButton',
    "exercicio_dinamico": 'org.grapentin.apps.exceer:id/TrainingActivityCurrentExerciseLabel',
    "pause": 'org.grapentin.apps.exceer:id/TrainingActivityContextButton',
    "agachamento": 'org.grapentin.apps.exceer:id/TrainingActivityCurrentExerciseLabel',
    "series": 'org.grapentin.apps.exceer:id/TrainingActivityProgressLabel',
    "alerta": 'android:id/message',
    "botao nao": 'android:id/button2',
    "botao sim": 'android:id/button1',
    "flexao_pernas": 'org.grapentin.apps.exceer:id/TrainingActivityCurrentExerciseLabel',
    "flexao_maos": 'org.grapentin.apps.exceer:id/TrainingActivityCurrentExerciseLabel',
    "flexao_parede": 'org.grapentin.apps.exceer:id/TrainingActivityCurrentExerciseLabel',
    "horizontal_pull": 'org.grapentin.apps.exceer:id/TrainingActivityCurrentExerciseLabel',
    "prancha": 'org.grapentin.apps.exceer:id/TrainingActivityCurrentExerciseLabel'
}
```

```python
dataframes = []

for historia in historias_usuario:
    for campo, id_aplicacao in mapeamento_campos.items():
        match = None
        valor = None
        action = "type"
        tipo = "edittext"

        if "ícone" in campo:
            match = re.search(regex_icone, historia)
        elif "start now" in campo:
            match = re.search(regex_start_now, historia)
            action, tipo = "click", "button"
        elif "start" in campo and "now" not in campo:
            match = re.search(regex_start, historia)
            action, tipo = "click", "button"
        elif "exercicio_dinamico" in campo:
            match = re.search(regex_exercicio_dinamico, historia)
        elif "pause" in campo:
            match = re.search(regex_pause, historia)
            action, tipo = "click", "button"
        elif "agachamento" in campo:
            match = re.search(regex_agachamento, historia)
        elif "series" in campo:
            match = re.search(regex_series, historia)
        elif "alerta" in campo:
            match = re.search(regex_alerta, historia)
        elif "botao nao" in campo:
            match = re.search(regex_botao_nao, historia)
            action, tipo = "click", "button"
        elif "botao sim" in campo:
            match = re.search(regex_botao_sim, historia)
            action, tipo = "click", "button"
        elif "flexao_pernas" in campo:
            match = re.search(regex_flexao_pernas, historia)
        elif "flexao_maos" in campo:
            match = re.search(regex_flexao_maos, historia)
        elif "flexao_parede" in campo:
            match = re.search(regex_flexao_parede, historia)
        elif "horizontal_pull" in campo:
            match = re.search(regex_horizontal_pull, historia)
        elif "prancha" in campo:
            match = re.search(regex_prancha, historia)

        if match:
            valor = match.group(0).strip("'") if match.lastindex is not None and match.lastindex >= 1 else match.group().strip()

            dataframes.append(pd.DataFrame({
                "Activity": ["App activity"],
                "Field": [campo],
                "ID": [id_aplicacao],
                "Action": [action],
                "Type": [tipo],
                "Size_Start": [1],
                "Size_End": [len(valor)],
                "Value": [valor]
            }))

if dataframes:
    df_final = pd.concat(dataframes, ignore_index=True)
    df_final.to_csv("Exceer.csv", index=False)
else:
    pass
```
