```python
import pandas as pd
import re
```

```python
historias_usuario = [
    "Como usuário, eu quero clicar no ícone do Dev BetterCounter para iniciar o aplicativo facilmente.",
    "Como usuário, eu quero criar uma nova atividade, como 'Estudo de Geografia', para registrar sua frequência.",
    "Como usuário, eu quero poder aumentar ou diminuir a quantidade de vezes que realizei a atividade 'Estudo de Geografia', para manter um registro preciso e atualizado das minhas atividades.",
    "Como usuário, eu quero poder visualizar um gráfico que mostra a frequência com que realizei a atividade 'Estudo de Geografia', para avaliar meu progresso e comprometimento com essa atividade.",
    "Como usuário, eu quero criar uma atividade para registrar quantas vezes pratiquei exercício físico, para monitorar minha rotina de atividades físicas.",
    "Como usuário, eu quero poder ajustar a quantidade de vezes que pratiquei exercício físico, para manter um registro atualizado das minhas atividades físicas.",
    "Como usuário, eu quero poder visualizar um gráfico que mostra a frequência com que pratiquei exercício físico, para acompanhar meu progresso na rotina de exercícios físicos.",
    "Como usuário, eu quero criar uma atividade para registrar quantas vezes li um livro, para monitorar minha rotina de leitura.",
    "Como usuário, eu quero poder ajustar a quantidade de vezes que li um livro, para manter um registro atualizado das minhas leituras.",
    "Como usuário, eu quero poder visualizar um gráfico que mostra a frequência com que li um livro, para acompanhar meu progresso na rotina de leitura.",
    "Como usuário, eu quero criar uma atividade para registrar quantas vezes pratiquei meditação, para monitorar minha prática de mindfulness.",
    "Como usuário, eu quero poder ajustar a quantidade de vezes que pratiquei meditação, para manter um registro atualizado da minha prática de mindfulness.",
    "Como usuário, eu quero poder visualizar um gráfico que mostra a frequência com que pratiquei meditação, para acompanhar meu progresso na prática de mindfulness.",
    "Como usuário, eu quero criar uma atividade para registrar quantas vezes socializei com amigos, para monitorar minha interação social.",
    "Como usuário, eu quero poder ajustar a quantidade de vezes que socializei com amigos, para manter um registro atualizado da minha interação social.",
    "Como usuário, eu quero poder visualizar um gráfico que mostra a frequência com que socializei com amigos, para acompanhar meu progresso na interação social."
]
```

```python
regex_ícone = r"ícone do Dev BetterCounter"
regex_adicionar = r"criar uma nova atividade, como '(.+?)',"
regex_campo_de_texto = r"quantas vezes (.+?),"
regex_cancelar = r"cancelar"
regex_adicionar_id = r"criar uma atividade"
regex_task = r"para (.+?),"
regex_incremento = r"aumentar ou diminuir a quantidade de vezes que (.+?),"
regex_decremento = r"aumentar ou diminuir a quantidade de vezes que (.+?),"
regex_visualização = r"visualizar um gráfico que mostra a frequência com que (.+?),"
regex_gráfico = r"visualizar um gráfico que mostra a frequência com que (.+?),"

mapeamento_campos = {
    "ícone": 'adb shell pm list packages | findstr org.kde.bettercounter',
    "adicionar": 'accessibility id Add counter',
    "campo de texto": 'id org.kde.bettercounter.debug:id/editText',
    "cancelar": 'id android:id/button2',
    "adicionar_id": 'id android:id/button1',
    "task": 'xpath /hierarchy/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.view.ViewGroup/androidx.recyclerview.widget.RecyclerView/android.widget.LinearLayout/android.widget.LinearLayout',
    "incremento": 'accessibility id Increase',
    "decremento": 'accessibility id Decrease',
    "visualização": 'id org.kde.bettercounter.debug:id/timestampText',
    "gráfico": 'id org.kde.bettercounter.debug:id/chart'
}
```

```python
dataframes = []

for historia in historias_usuario:
    for campo, id_aplicacao in mapeamento_campos.items():
        match = None
        valor = None

        if "ícone" in campo:
            match = re.search(regex_ícone, historia)
        elif "adicionar" in campo:
            match = re.search(regex_adicionar, historia)
        elif "campo de texto" in campo:
            match = re.search(regex_campo_de_texto, historia)
        elif "cancelar" in campo:
            match = re.search(regex_cancelar, historia)
        elif "adicionar_id" in campo:
            match = re.search(regex_adicionar_id, historia)
        elif "task" in campo:
            match = re.search(regex_task, historia)
        elif "incremento" in campo:
            match = re.search(regex_incremento, historia)
        elif "decremento" in campo:
            match = re.search(regex_decremento, historia)
        elif "visualização" in campo:
            match = re.search(regex_visualização, historia)
        elif "gráfico" in campo:
            match = re.search(regex_gráfico, historia)

        if match:
            valor = match.group(1).strip("'") if match.lastindex is not None and match.lastindex >= 1 else match.group().strip()


            if "ícone" in campo or "cancelar" in campo or "adicionar_id" in campo or "incremento" in campo or "decremento" in campo:
                action_type = "button"
            else:
                action_type = "edittext"

            dataframes.append(pd.DataFrame({
                "Activity": ["App activity"],
                "Field": [campo.split()[-1]],
                "ID": [id_aplicacao],
                "Action": ["click" if action_type == "button" else "type"],
                "Type": [action_type],
                "Size_Start": [1],
                "Size_End": [len(valor)],
                "Value": [valor]
            }))

if dataframes:
    df_final = pd.concat(dataframes, ignore_index=True)
    df_final.to_csv("dev.csv", index=False)
else:
    pass
```
