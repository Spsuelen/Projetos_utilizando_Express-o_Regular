```python
import pandas as pd
import re
```

```python
historias_usuario = [
    "Como usuário, quero encontrar e clicar no ícone do Simple Text para abrir o editor de texto.",
    "Como usuário, quero preencher o campo de texto 'criar a tarefa \'Estudar Matemática\'' para organizar meu tempo de estudo.",
    "Como usuário, quero usar o teclado para digitar informações na tela de criação de tarefa.",
    "Como usuário, quero acessar o menu de contexto para realizar ações adicionais na interface do aplicativo.",
    "Como usuário, quero inserir o nome do arquivo na caixa de texto 'nome do arquivo' para salvar minhas alterações.",
    "Como usuário, quero ver o relatório semanal na lista de relatórios para revisar meu progresso.",
    "Como usuário, quero clicar no botão 'salvar' para confirmar minhas alterações.",
    "Como usuário, quero cancelar uma ação clicando no botão 'cancelar'.",
    "Como usuário, quero receber um alerta se tentar criar uma tarefa com um nome já existente ('alerta nomes iguais').",
    "Como usuário, quero confirmar uma ação clicando no botão 'confirmar'.",
    "Como usuário, quero descartar uma ação clicando no botão 'excluir' ou 'descartar'.",
    "Como usuário, quero criar um novo item selecionando 'Novo' na lista de opções disponíveis.",
    "Como usuário, quero abrir um item selecionando 'Abrir' na lista de opções disponíveis.",
    "Como usuário, quero salvar um item selecionando 'salvar' na lista de opções disponíveis.",
    "Como usuário, quero selecionar 'salvar como' na lista de opções disponíveis para salvar um item com um novo nome.",
    "Como usuário, quero visualizar o registro de tempo ao selecionar 'visualizar o registro de tempo' na lista de opções disponíveis.",
    "Como usuário, quero acessar as configurações do aplicativo ao selecionar 'alterar as opções de configuração' na lista de opções disponíveis.",
    "Como usuário, quero sair do aplicativo ao selecionar 'sair' na lista de opções disponíveis."
]
```

```python
regex_icone_simple_text = r"ícone do Simple Text"
regex_campo_texto = r"(?:criar|iniciar|parar|editar) a tarefa '([^']*)'"
regex_teclado = r"teclado"
regex_tres_pontinhos = r"menu de contexto"
regex_nome_arquivo = r"nome do arquivo"
regex_lista_texto = r"relatório semanal"
regex_salvar = r"salvar"
regex_cancelar = r"cancelar"
regex_alerta_nomes_iguais = r"alerta nomes iguais"
regex_confirmar = r"confirmar"
regex_descartar = r"excluir|descartar"
regex_new = r"Novo"
regex_open = r"Abrir"
regex_save = r"salvar"
regex_save_as = r"salvar"
regex_search = r"visualizar o registro de tempo"
regex_setting = r"alterar as opções de configuração"
regex_exit = r"sair"

mapeamento_campos = {
    "ícone do Simple Text": 'Simple Text Editor (accessibility id)',
    "campo de texto": 'com.maxistar.textpad:id/editText1',
    "teclado": 'android:id/decor_content_parent',
    "menu de contexto": 'More options',
    "nome do arquivo": 'com.maxistar.textpad:id/fdEditTextFile',
    "relatório semanal": 'android:id/list',
    "salvar": 'com.maxistar.textpad:id/fdButtonCreate',
    "cancelar": 'com.maxistar.textpad:id/fdButtonCancel',
    "alerta nomes iguais": '/hierarchy/android.widget.FrameLayout',
    "confirmar": 'android:id/button1',
    "descartar": 'android:id/button2',
    "Novo": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[1]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "Abrir": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[2]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "salvar": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[3]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "salvar": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[4]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "visualizar o registro de tempo": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[5]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "alterar as opções de configuração": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[6]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
    "sair": '/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[7]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView',
}
```

```python
dataframes = []


for historia in historias_usuario:
    for campo, id_aplicacao in mapeamento_campos.items():
        match = None
        valor = None

        if "ícone do Simple Text" in campo:
            match = re.search(regex_icone_simple_text, historia)
        elif "campo de texto" in campo:
            match = re.search(regex_campo_texto, historia)
        elif "teclado" in campo:
            match = re.search(regex_teclado, historia)
        elif "menu de contexto" in campo:
            match = re.search(regex_tres_pontinhos, historia)
        elif "nome do arquivo" in campo:
            match = re.search(regex_nome_arquivo, historia)
        elif "relatório semanal" in campo:
            match = re.search(regex_lista_texto, historia)
        elif "salvar" in campo:
            match = re.search(regex_salvar, historia)
        elif "cancelar" in campo:
            match = re.search(regex_cancelar, historia)
        elif "alerta nomes iguais" in campo:
            match = re.search(regex_alerta_nomes_iguais, historia)
        elif "confirmar" in campo:
            match = re.search(regex_confirmar, historia)
        elif "descartar" in campo:
            match = re.search(regex_descartar, historia)
        elif "Novo" in campo:
            match = re.search(regex_new, historia)
        elif "Abrir" in campo:
            match = re.search(regex_open, historia)
        elif "salvar" in campo:
            match = re.search(regex_save, historia)
        elif "visualizar o registro de tempo" in campo:
            match = re.search(regex_search, historia)
        elif "alterar as opções de configuração" in campo:
            match = re.search(regex_setting, historia)
        elif "sair" in campo:
            match = re.search(regex_exit, historia)

        if match:
            valor = match.group(1).strip() if match.lastindex is not None else match.group().strip()

        if valor:

            dataframes.append(pd.DataFrame({
                "Activity": ["App activity"],
                "Field": [campo],
                "ID": [id_aplicacao],
                "Action": ["click" if "botão" in campo else "type"],
                "Type": ["button" if "botão" in campo else "edittext"],
                "Size_Start": [1],
                "Size_End": [len(valor)],
                "Value": [valor]
            }))


if dataframes:
    df_final = pd.concat(dataframes, ignore_index=True)
    df_final.to_csv("Atime.csv", index=False)
else:
    pass
```
