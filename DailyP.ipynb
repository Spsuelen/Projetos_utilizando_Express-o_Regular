```python
historias_usuario = [
    "Como usuário, quero ver o ícone 'Daily Pill' na tela inicial para abrir o aplicativo facilmente.",
    "Como usuário, quero definir um lembrete para me lembrar de tomar meu remédio não tomado.",
    "Como usuário, quero confirmar a criação do lembrete ao pressionar o botão de confirmação.",
    "Como usuário, quero ativar o lembrete ao tocar no banner.",
    "Como usuário, quero escolher uma opção específica para ativar no aplicativo.",
    "Como usuário, quero configurar a hora do lembrete usando o seletor de horário automático.",
    "Como usuário, quero mudar para o modo de entrada manual para definir o horário do lembrete.",
    "Como usuário, quero inserir a hora manualmente no campo de entrada de hora.",
    "Como usuário, quero definir os minutos manualmente no campo de entrada de minutos.",
    "Como usuário, quero selecionar o turno AM ao clicar na opção correspondente.",
    "Como usuário, quero escolher o turno PM ao clicar na opção correspondente.",
    "Como usuário, quero cancelar a configuração do lembrete ao pressionar o botão de cancelamento.",
    "Como usuário, quero salvar as configurações do lembrete ao pressionar o botão de confirmação.",
    "Como usuário, quero acessar a opção de bloqueio ao interagir com o elemento específico.",
    "Como usuário, quero voltar para a tela anterior ao pressionar o botão 'Voltar'."
]
```

```python
regex_icone = r"ícone\s+(?:=|é)\s+'(.+?)'"
regex_mensagem_lembrete = r"definir um lembrete\s+(?:para|de)\s+(.+)"
regex_confirmar = r"confirmar a criação do lembrete\s+(?:ao|ao apertar|apertar|pressionar|pressionar o botão)\s+(.+)"
regex_botao_ativar_lembrete = r"ativar o lembrete\s+(?:ao|ao tocar|tocar|no banner)\s+(.+)"
regex_opcao_ativar = r"escolher uma opção específica\s+(?:para|para ativar)\s+(.+)"
regex_hora_lembrete = r"configurar a hora do lembrete\s+(?:usando|usando o seletor|o seletor)\s+(.+)"
regex_hora_automatica = r"mudar para o modo de entrada manual\s+(?:para definir|para definir o horário)\s+(.+)"
regex_hora_manual = r"inserir a hora manualmente\s+(?:no campo de entrada de hora|no campo de entrada de horas|campo de entrada de horas)\s+(.+)"
regex_campo_hora_manual = r"definir os minutos manualmente\s+(?:no campo de entrada de minutos|campo de entrada de minutos)\s+(.+)"
regex_campo_minto_manual = r"selecionar o turno da manhã\s+(?:ao clicar|clicar)\s+(.+)"
regex_seletor_turno = r"escolher o turno da tarde\s+(?:ao clicar|clicar)\s+(.+)"
regex_turno_am = r"cancelar a configuração do lembrete\s+(?:ao pressionar|pressionar|pressionar o botão)\s+(.+)"
regex_turno_pm = r"salvar as configurações do lembrete\s+(?:ao pressionar|pressionar|pressionar o botão)\s+(.+)"
regex_botao_cancelar = r"acessar a opção de bloqueio\s+(?:ao interagir|interagir|com o elemento específico)\s+(.+)"
regex_botao_ok = r"voltar para a tela anterior\s+(?:ao pressionar|pressionar)\s+(.+)"
regex_botao_bloqueio = r"ver o ícone 'Daily Pill'\s+(?:na tela inicial|para abrir o aplicativo facilmente)\s+(.+)"
regex_botao_teste_lembrete = r"acessar a opção de configuração\s+(?:ao interagir|interagir|com o elemento específico)\s+(.+)"
regex_botao_voltar = r"acessar a opção de configuração\s+(?:ao interagir|interagir|com o elemento específico)\s+(.+)"
regex_tema = r"escolher o tema\s+(?:ao clicar|clicar)\s+(.+)"
regex_tema_claro = r"escolher o tema claro\s+(?:ao clicar|clicar)\s+(.+)"
regex_tema_escuro = r"escolher o tema escuro\s+(?:ao clicar|clicar)\s+(.+)"
regex_tema_automatico = r"escolher o tema automático\s+(?:ao clicar|clicar)\s+(.+)"
regex_botao_resetar = r"acessar o botão de reset\s+(?:ao interagir|interagir|com o elemento específico)\s+(.+)"
regex_opcoes = r"acessar mais opções\s+(?:ao interagir|interagir|com o elemento específico)\s+(.+)"
regex_opcao_configuracao = r"acessar a opção de configuração\s+(?:ao interagir|interagir|com o elemento específico)\s+(.+)"
regex_opcao_sobre = r"acessar a opção Sobre\s+(?:ao interagir|interagir|com o elemento específico)\s+(.+)"



mapeamento_elementos = {
    "icone": "accessibility id	Daily Pill",
    "mensagem_lembrete": "id	br.com.gualandi.dailypill:id/not_taken_message",
    "confirmar": "id	br.com.gualandi.dailypill:id/button",
    "botao_ativar_lembrete": "id	br.com.gualandi.dailypill:id/banner_button",
    "opcao_ativar": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/androidx.recyclerview.widget.RecyclerView/android.widget.LinearLayout[2]",
    "hora_lembrete": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/androidx.recyclerview.widget.RecyclerView/android.widget.LinearLayout[3]/android.widget.RelativeLayout",
    "hora_automatica": "id	android:id/radial_picker",
    "hora_manual": "accessibility id	Switch to text input mode for the time input.",
    "campo_hora_manual": "id	android:id/input_hour",
    "campo_minto_manual": "id	android:id/input_minute",
    "seletor_turno": "id	android:id/text1",
    "turno_am": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.CheckedTextView[1]",
    "turno_pm": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.CheckedTextView[2]",
    "botao_cancelar": "id	android:id/button2",
    "botao_ok": "id	android:id/button1",
    "botao_bloqueio": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/androidx.recyclerview.widget.RecyclerView/android.widget.LinearLayout[4]",
    "botao_teste_lembrete": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/androidx.recyclerview.widget.RecyclerView/android.widget.LinearLayout[5]/android.widget.LinearLayout/android.widget.Button",
    "botao_voltar": "accessibility id	Navigate up",
    "tema": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/androidx.recyclerview.widget.RecyclerView/android.widget.LinearLayout[6]/android.widget.RelativeLayout",
    "tema_claro": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.FrameLayout/androidx.appcompat.widget.LinearLayoutCompat/android.widget.FrameLayout/android.widget.ListView/android.widget.CheckedTextView[1]",
    "tema_escuro": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.FrameLayout/androidx.appcompat.widget.LinearLayoutCompat/android.widget.FrameLayout/android.widget.ListView/android.widget.CheckedTextView[2]",
    "tema_automatico": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.FrameLayout/androidx.appcompat.widget.LinearLayoutCompat/android.widget.FrameLayout/android.widget.ListView/android.widget.CheckedTextView[3]",
    "botao_resetar": "accessibility id	Reset",
    "opcoes": "accessibility id	More options",
    "opcao_configuracao": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[1]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView",
    "opcao_sobre": "xpath	/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.ListView/android.widget.LinearLayout[2]/android.widget.LinearLayout/android.widget.RelativeLayout/android.widget.TextView"
}
```

```python
dataframes = []

for historia in historias_usuario:
    for campo, id_aplicacao in mapeamento_elementos.items():
        match = None
        valor = None
        action = "type"
        tipo = "edittext"

        if "icone" in campo:
            match = re.search(regex_icone, historia)
        elif "mensagem_lembrete" in campo:
            match = re.search(regex_mensagem_lembrete, historia)
        elif "confirmar" in campo:
            match = re.search(regex_confirmar, historia)
        elif "botao_ativar_lembrete" in campo:
            match = re.search(regex_botao_ativar_lembrete, historia)
        elif "opcao_ativar" in campo:
            match = re.search(regex_opcao_ativar, historia)
        elif "hora_lembrete" in campo:
            match = re.search(regex_hora_lembrete, historia)
        elif "hora_automatica" in campo:
            match = re.search(regex_hora_automatica, historia)
        elif "hora_manual" in campo:
            match = re.search(regex_hora_manual, historia)
        elif "campo_hora_manual" in campo:
            match = re.search(regex_campo_hora_manual, historia)
        elif "campo_minto_manual" in campo:
            match = re.search(regex_campo_minto_manual, historia)
        elif "seletor_turno" in campo:
            match = re.search(regex_seletor_turno, historia)
        elif "turno_am" in campo:
            match = re.search(regex_turno_am, historia)
        elif "turno_pm" in campo:
            match = re.search(regex_turno_pm, historia)
        elif "botao_cancelar" in campo:
            match = re.search(regex_botao_cancelar, historia)
            action, tipo = "click", "button"
        elif "botao_ok" in campo:
            match = re.search(regex_botao_ok, historia)
            action, tipo = "click", "button"
        elif "botao_bloqueio" in campo:
            match = re.search(regex_botao_bloqueio, historia)
        elif "botao_teste_lembrete" in campo:
            match = re.search(regex_botao_teste_lembrete, historia)
        elif "botao_voltar" in campo:
            match = re.search(regex_botao_voltar, historia)
        elif "tema" in campo:
            match = re.search(regex_tema, historia)
        elif "tema_claro" in campo:
            match = re.search(regex_tema_claro, historia)
        elif "tema_escuro" in campo:
            match = re.search(regex_tema_escuro, historia)
        elif "tema_automatico" in campo:
            match = re.search(regex_tema_automatico, historia)
        elif "botao_resetar" in campo:
            match = re.search(regex_botao_resetar, historia)
        elif "opcoes" in campo:
            match = re.search(regex_opcoes, historia)
        elif "opcao_configuracao" in campo:
            match = re.search(regex_opcao_configuracao, historia)
        elif "opcao_sobre" in campo:
            match = re.search(regex_opcao_sobre, historia)

        if match:
            valor = match.group(1).strip("'") if match.lastindex is not None and match.lastindex >= 1 else match.group().strip()

            dataframes.append(pd.DataFrame({
                "Activity": ["App activity"],
                "Field": [campo],
                "ID": [id_aplicacao],
                "Action": [action],
                "Type": [tipo],
                "Size_Start": [1],
                "Size_End": [len(valor)],
                "Value": [valor]
            }))
if dataframes:
    df_final = pd.concat(dataframes, ignore_index=True)
    df_final.to_csv("DailyPill.csv", index=False)
else:
    pass
```
